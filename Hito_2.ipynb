## steam

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

steam = pd.read_csv('steam.csv')
steam.tail()
```

### Tabla GENERO

```python
genero = steam.groupby(['genres'])
genero = genero.size().reset_index(name='count')

# Separamos los géneros
genero['genres'] = genero['genres'].str.split(';')
genero = genero.explode('genres')
genero = genero.groupby(['genres']).sum().reset_index()

# Eliminamos el conteo
genero = genero.drop(['count'], axis=1)
genero

```

### Tabla CATEGORIES

```python
categories = steam.groupby(['categories'])
categories = categories.size().reset_index(name='count')

# Separamos las categorías
categories['categories'] = categories['categories'].str.split(';')
categories = categories.explode('categories')
categories = categories.groupby(['categories']).sum().reset_index()

# Eliminamos el conteo
categories = categories.drop(['count'], axis=1)
categories
```

### Tabla PLATFORMS

```python
platforms = steam.groupby(['platforms'])
platforms = platforms.size().reset_index(name='count')

# Separamos las plataformas
platforms['platforms'] = platforms['platforms'].str.split(';')
platforms = platforms.explode('platforms')
platforms = platforms.groupby(['platforms']).sum().reset_index()

# Eliminamos el conteo
platforms = platforms.drop(['count'], axis=1)
platforms
```

### Tabla STEAM_SPYTAGS

```python
steam_spytags = steam.groupby(['steamspy_tags'])
steam_spytags = steam_spytags.size().reset_index(name='count')

# Separamos los spy_tags
steam_spytags['steamspy_tags'] = steam_spytags['steamspy_tags'].str.split(';')
steam_spytags = steam_spytags.explode('steamspy_tags')
steam_spytags['steamspy_tags'].unique()
steam_spytags = steam_spytags.groupby(['steamspy_tags']).sum().reset_index()

# Eliminamos el conteo
steam_spytags = steam_spytags.drop(['count'], axis=1)
steam_spytags
```

## steam_info

```python
media_data = pd.read_csv('steam_media_data.csv')
desc_data = pd.read_csv('steam_description_data.csv')
```

```python
# Eliminamos las columnas que no nos sirven
media_data = media_data.drop(['screenshots', 'movies', 'background'], axis=1)
media_data.head()
```

```python
desc_data.head()
```

```python
# Union de los dataframes según el id
steam_info = pd.merge(desc_data, media_data, on='steam_appid', how='left')
steam_info.head()
```

```python
# Buscar nulls
steam_info.isnull().sum()

# Cambiamos los nulls por 0
steam_info = steam_info.fillna(0)
```

## relations

### Género y juegos

```python
# Añadimos un id a genero en la primera columna
genero.insert(0, 'genre_id', range(0, 0 + len(genero)))
```

```python
# Creamos un dataframe con la id de los juegos que contengan el género
import warnings
warnings.filterwarnings('ignore')
genre_id = pd.DataFrame(columns=['genre_id', 'steam_appid'])
for i in range(len(steam)):
    if i % 100 == 0:
        print(i)
    for j in range(len(genero)):
        if genero['genres'][j] in steam['genres'][i]:
            genre_id = genre_id.append({'genre_id': genero['genre_id'][j], 'steam_appid': steam['appid'][i]}, ignore_index=True)
```

```python
# Pasamos a int
genre_id['genre_id'] = genre_id['genre_id'].astype(int)
genre_id['steam_appid'] = genre_id['steam_appid'].astype(int)
```

### Categoría y juego

```python
# Añadimos un id a categories en la primera columna
categories.insert(0, 'category_id', range(0, 0 + len(categories)))

# Creamos un dataframe con la id de los juegos que contengan la categoría
category_id = pd.DataFrame(columns=['category_id', 'steam_appid'])
for i in range(len(steam)):
    if i % 100 == 0:
        print(i)
    for j in range(len(categories)):
        if categories['categories'][j] in steam['categories'][i]:
            category_id = category_id.append({'category_id': categories['category_id'][j], 'steam_appid': steam['appid'][i]}, ignore_index=True)

# Pasamos a int
category_id['category_id'] = category_id['category_id'].astype(int)
category_id['steam_appid'] = category_id['steam_appid'].astype(int)
```

### Plataforma y juego

```python
# Añadimos un id a platforms en la primera columna
platforms.insert(0, 'platform_id', range(0, 0 + len(platforms)))

# Creamos un dataframe con la id de los juegos que contengan la plataforma
platform_id = pd.DataFrame(columns=['platform_id', 'steam_appid'])
for i in range(len(steam)):
    if i % 100 == 0:
        print(i)
    for j in range(len(platforms)):
        if platforms['platforms'][j] in steam['platforms'][i]:
            platform_id = platform_id.append({'platform_id': platforms['platform_id'][j], 'steam_appid': steam['appid'][i]}, ignore_index=True)

# Pasamos a int
platform_id['platform_id'] = platform_id['platform_id'].astype(int)
platform_id['steam_appid'] = platform_id['steam_appid'].astype(int)
```

### SteamSpyTags y juego

```python
# Añadimos un id a steam_spytags en la primera columna
steam_spytags.insert(0, 'tag_id', range(0, 0 + len(steam_spytags)))

# Creamos un dataframe con la id de los juegos que contengan el tag
tag_id = pd.DataFrame(columns=['tag_id', 'steam_appid'])
for i in range(len(steam)):
    if i % 100 == 0:
        print(i)
    for j in range(len(steam_spytags)):
        if steam_spytags['steamspy_tags'][j] in steam['steamspy_tags'][i]:
            tag_id = tag_id.append({'tag_id': steam_spytags['tag_id'][j], 'steam_appid': steam['appid'][i]}, ignore_index=True)

# Pasamos a int
tag_id['tag_id'] = tag_id['tag_id'].astype(int)
tag_id['steam_appid'] = tag_id['steam_appid'].astype(int)
```
